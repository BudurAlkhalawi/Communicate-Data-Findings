# Univariate Exploration

#### I'll start by looking at the distribution of the interest variable: arrival delay 

```python
# start with a standard-scaled plot
binsize = 60
bins = np.arange(0, mergedFrames['ArrDelay'].max()+binsize, binsize)

plt.figure(figsize=[14.70, 8.27])
plt.hist(data = mergedFrames, x = 'ArrDelay',bins=bins);
plt.xlim([0,400])
plt.ylabel('frequently')
plt.xlabel('arrival delay in minutes')
plt.title('distribution of arrival delay in minutes')
plt.show()
```

#### And I'll look at the distribution of the interest variable: departure delay 

```python
# plotting DepDelay on a standard scale
binsize = 60
bins = np.arange(0, mergedFrames['DepDelay'].max()+binsize, binsize)

plt.figure(figsize=[14.70, 8.27])
plt.hist(data = mergedFrames, x = 'DepDelay',bins=bins);
plt.xlim([0,400])
plt.ylabel('frequently')
plt.xlabel('departure delay in minutes')
plt.title('distribution of departure delay in minutes')
plt.show()
```

### The most frequent delay is one hour for departure and arrival

#### How many canceled flights compared to non-canceled?(1 = yes, 0 = no)

```python
# count plot for Cancelled
default_color = sb.color_palette()[0]
plt.figure(figsize=[14.70, 8.27])
plt.title('Distribution of Cancellation')
sb.countplot(data = mergedFrames.sample(1000000), x = 'Cancelled', color = default_color);
```

### I took a sample 1 million from the data and found that the difference between canceled and non-canceled flights is very large, and the number of canceled is small.

# Bivariate Exploration

### To start off with, I want to look at the pairwise correlations present between variables in the data.

#### I'll look at the relationship between distance and arrival time for each 2008 and 2007 alone.

```python
# scatter plot of Distance vs. ArrDelay
plt.figure(figsize=[14.70, 8.27])
plt.title('Relation between Distance and ArrDelay 2008')
sb.regplot(data = df_2008, x = 'Distance', y = 'ArrDelay', 
           x_jitter = 0.2, y_jitter = 0.2, scatter_kws = {'alpha' : 1/3});
```

```python
# scatter plot of Distance vs. ArrDelay
plt.figure(figsize=[14.70, 8.27])
plt.title('Relation between Distance and ArrDelay 2007')
sb.regplot(data = df_2007, x = 'Distance', y = 'ArrDelay', 
           x_jitter = 0.2, y_jitter = 0.2, scatter_kws = {'alpha' : 1/3});
```

#### It also shows from the distribution that the relationship is stronger between zero and 3000 miles for Distance with the arrival time from zero to 500 minutes in both 2007 and 2008 year.

### Now, let's take a second look at the CRSElapsedTime distribution on the categorical variables: CancellationCode in 2007.

```python
# convert 'A','B','C' and 'D' into ordered categorical types
ordinal_var = ['A','B','C','D']
ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                                categories = ordinal_var)
df_2007['CancellationCode'] = df_2007['CancellationCode'].astype(ordered_var)
base_color = sb.color_palette()[0]

# plot the categorical variable(CancellationCode)and CRSArrTime
plt.figure(figsize=[14.70, 8.27])
plt.title('Relation between CancellationCode and CRSArrTime')
sb.violinplot(data = df_2007, x = 'CancellationCode', y = 'CRSArrTime',
                  color = base_color);
```

#### The distribution for each of the reasons for the abolition takes a different form, C(NAS) and D(security) almost have the same distribution medium.

# Multivariate Exploration

### The main thing I want to explore in this part of the analysis is how the categorical variable (CancellationCode) plays into the relationship between the Month and the top 10 destinations.
* Knowing that reason for cancellation (A = carrier, B = weather, C = NAS, D = security) 

```python
# preparation data, choose the top 10 destinations.
df_2007.Dest.value_counts().head(10)
```

```python
top_10_Dest = ['ATL','ORD','DFW','DEN','LAX','PHX','IAH','LAS','DTW','MSP']
```

```python
temp_df = df_2007[df_2007['Dest'].isin(top_10_Dest)]
```

```python
temp_df.Dest.value_counts()
```

```python
# plot the categorical variable (CancellationCode) with Month and destinations
plt.figure(figsize=[14.70, 8.27])
plt.title('Relation between Distance and Month with CancellationCode')
sb.stripplot(x="Dest", y="Month", hue="CancellationCode",
                    data=temp_df, jitter=True,alpha=.3,
                 palette="Set1", dodge=True, size=8);
```

#### From this plot, I conclude that the lowest reasons for cancellation are security(D)
#### And the least times that a cancellation occurred due to (NAS) was in November and at the destination DEN
#### And the least times that a cancellation occurred due to (weather) was in October and at the destination LAS
